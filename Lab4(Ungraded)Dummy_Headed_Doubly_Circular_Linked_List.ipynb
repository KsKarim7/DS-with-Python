```python
class Node:
  def __init__(self, e, n, p):
    self.elem = e
    self.next = n
    self.prev = p
```

```python
class DoublyList:

  # Creates a Dummy Head for Dummy Headed Circular Doubly Circular Linked List.
  def __init__(self):
    # To Do
    # pass # Remove this line
    self.dh = Node(None,None,None)
    self.dh.next =   self.dh
    self.dh.prev=   self.dh
    self.tail =   self.dh


  #create a DHDCL from an array
  def create_list(self, l):
    # To Do
    # pass # Remove this line
    for i in range(len(l)):
        n = Node(l[i], self.dh,self.tail)
        self.tail.next = n
        self.tail = self.tail.next
        self.dh.prev = self.tail


  # Counts the number of Nodes in the list and return the number
  def countNode(self):
    # To Do
    # pass # Remove this line
    temp = self.dh.next
    count = 0
    while(temp != self.dh):
        count+=1
        temp = temp.next
    return count


  # prints the elements in the list forward
  def forwardprint(self):
    # To Do
    # pass # Remove this line
    temp = self.dh.next
    while(temp != self.dh):
        print(temp.elem,end=", ")
        temp = temp.next

  # prints the elements in the list backward
  def backwardprint(self):
    # To Do
    # pass # Remove this line
    temp = self.tail
    while(temp != self.dh):
        print(temp.elem,end= ', ')
        temp = temp.prev

  # returns the reference of the at the given index. For invalid index return None.
  def nodeAt(self, idx):
    # To Do
    # pass # Remove this line
    temp = self.dh.next
    count = 0
    while(temp != self.dh):
        if(count == idx):
            return temp
        count+=1
        temp = temp.next
    return None


  # returns the index of the containing the given element. if the element does not exist in the List, return -1.
  def indexOf(self, elem):
    # To Do
    # pass # Remove this line
    temp = self.dh.next
    count = 0
    flag = False
    while(temp != self.dh):
        if(temp.elem == elem):
            flag = True
            break
        temp = temp.next
        count +=1
    if(flag == True):
        return count
    else: return '-1'



  # insert at specific index of the DHDCL
  def insert(self, elem, idx):
    # To Do
    # pass # Remove this line
    ins = Node(elem,None,None)
    if(idx == self.countNode()):
        ins.next = self.dh
        ins.prev = self.tail
        self.tail.next = ins
        self.dh.prev = ins
        self.tail = ins


        # print('where am i?',self.countNode())
        # print()
        # n= self.nodeAt(self.countNode()-2)
        # print(n.elem,'idx-',self.countNode()-2)
        # print(self.countNode())
        # print(last.prev.elem,"-- here")
    else:
        current = self.nodeAt(idx)
        prevNode = current.prev
        current.prev = ins
        prevNode.next = ins
        ins.next = current
        ins.prev = prevNode


  #removes the node of the specific index and returns the element of the node
  def remove(self, idx):
    # To Do
    # pass # Remove this line
    remove = self.nodeAt(idx)

    prev = remove.prev
    next = remove.next

    prev.next = next
    next.prev = prev
    if(idx == self.countNode()): self.tail = prev
    return remove.elem
```
