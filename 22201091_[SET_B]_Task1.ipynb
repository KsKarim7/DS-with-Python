# Solver code

```python

def prime_count(n, num):
    if num == 1:
        return 1

    if n % num == 0:
        return 0

    return prime_count(n, num-1)

def countPrime(n):
    if n == 1:
        return 0
    val = prime_count(n, n-1)
    return val + countPrime(n-1)
```

# Driver Code

```python
import sys
sys.setrecursionlimit(99999)
primeFlag = [1] * 999
primeFlag[1] = 0
for itr in range(2, 999):
    if primeFlag[itr] == 1:
        for jtr in range(itr + itr, 999, itr):
           primeFlag[jtr] = 0
    primeFlag[itr] += primeFlag[itr - 1]
flag = 0
for itr in range(1, 999):
    if primeFlag[itr] != countPrime(itr):
        print('Your solution is wrong for n =', itr)
        flag = 1
        break
if flag == 0:
    print('Accepted')
```
