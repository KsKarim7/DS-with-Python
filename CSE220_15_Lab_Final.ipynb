```python
class BTNode:
  def __init__(self, elem):
    self.elem = elem
    self.right = None
    self.left = None
```

```python
#Input 1
root1 = BTNode(9)
node2 = BTNode(4)
node3 = BTNode(5)
node4 = BTNode(18)
node5 = BTNode(14)
node6 = BTNode(3)
node7 = BTNode(54)
node8 = BTNode(12)
node9 = BTNode(8)
node10 = BTNode(91)
node11 = BTNode(56)

root1.left = node2
root1.right = node3

node2.left = node4

node3.left = node5
node3.right = node6

node4.left = node7
node4.right = node8

node5.left = node9

node8.left = node10
node8.right = node11
```

```python
#Input 2
root2 = BTNode(2)
node12 = BTNode(4)
node13 = BTNode(5)
node14 = BTNode(9)
node15 = BTNode(11)
node16 = BTNode(98)
node17 = BTNode(62)
node18 = BTNode(17)
node19 = BTNode(15)
node20 = BTNode(45)
node21 = BTNode(8)

root2.left = node12
root2.right = node13

node12.left = node14

node13.left = node15
node13.right = node16

node14.left = node17
node14.right = node18

node18.left = node19
node18.right = node20

node20.right = node21
```

```python
def postorder(root):

    if(root == None):
        return
    postorder(root.left)
    postorder(root.right)
    print(root.elem,end=" ")

def modSum(root,lvl):
    if(root==None):
        return 0
    if(lvl == 0):
        return root.elem + modSum(root.left,lvl+1) + modSum(root.right,lvl+1)
    # modSum(root.left,lvl+1)
    # modSum(root.right,lvl+1)
    else:
        return root.elem%lvl + modSum(root.left,lvl+1) + modSum(root.right,lvl+1)

def sumTree(root):
    lvl = 0
    return modSum(root,lvl)



```

```python
postorder(root1)
postorder(root2)
print()
print(sumTree(root1))
print(sumTree(root2))
```
